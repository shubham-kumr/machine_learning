**Checking Unique and Number of unique values in each column**

```python
# Get unique values and their counts for each column
unique_values = {col: {'n_unique': df[col].nunique(), 'unique_values': df[col].unique()} for col in df.columns}

# Display the results
for col, values in unique_values.items():
    print(f"Column: {col} \n")
    print(f"Number of Unique Values: {values['n_unique']} \n")
    print(f"Unique Values: {values['unique_values']} \n")
```

# **Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Univariate Analysis**

1. Analysis of frequency of data distribution

```python
# Plot histograms for each numerical column
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 5, i + 1)  # Adjust subplot dimensions based on the number of columns
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

2. Analysis of outliers

```python
# Plot box plots for each numerical column
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 5, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```
