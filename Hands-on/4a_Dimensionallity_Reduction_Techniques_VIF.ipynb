# **Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Univariate Analysis**

1. Analysis of frequency of data distribution

```python
# Plot histograms for each numerical column
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 5, i + 1)  # Adjust subplot dimensions based on the number of columns
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

2. Analysis of outliers

```python
# Plot box plots for each numerical column
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 5, i + 1)
    sns.boxplot(x=df[col])
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```

**Bivariate analysis**

```python
# Plot scatter plots for each feature against MEDV
features = df.columns[:-1]  # Exclude MEDV as it's the target column

plt.figure(figsize=(15, 10))
for i, col in enumerate(features):
    plt.subplot(3, 5, i + 1)  # Adjust subplot dimensions as needed
    sns.scatterplot(x=df[col], y=df['MEDV'])
    plt.title(f'{col} vs MEDV')
    plt.xlabel(col)
    plt.ylabel('MEDV')
plt.tight_layout()
plt.show()
```

**Multivariate analysis**

```python
# Create the pair plot
sns.pairplot(df, hue='MEDV', palette='coolwarm', height=2.5)
plt.suptitle('Pair Plot of Features with MEDV as Hue', y=1.02)
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plotting the correlation of features with target column

correlation = df.drop('MEDV', axis=1).corrwith(df['MEDV'])

plt.figure(figsize=(10, 6))
sns.barplot(x=correlation.index, y=correlation.values, palette='coolwarm')
plt.title('Correlation of Features with MEDV')
plt.xlabel('Features')
plt.ylabel('Correlation with MEDV')
plt.xticks(rotation=45)
plt.axhline(0, color='grey', linestyle='--')
plt.show()
```

**Checking Multi-Collinearity Between Features**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import numpy as np
```

```python
# Define the features (excluding the target column 'MEDV')
features = df.drop('MEDV', axis=1)
```

```python
# Calculate VIF
vif_data = pd.DataFrame()
vif_data['Feature'] = features.columns
vif_data['VIF'] = [variance_inflation_factor(features.values, i) for i in range(features.shape[1])]

# Display VIF
vif_data.sort_values(by=['VIF'])
```
