**Checking Unique and Number of unique values in each column**

```python
# Get unique values and their counts for each column
unique_values = {col: {'n_unique': df[col].nunique(), 'unique_values': df[col].unique()} for col in df.columns}

# Display the results
for col, values in unique_values.items():
    print(f"Column: {col} \n")
    print(f"Number of Unique Values: {values['n_unique']} \n")
    print(f"Unique Values: {values['unique_values']} \n")
```
